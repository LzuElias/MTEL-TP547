```python
# Questão 1

import math

def calcular_probabilidade_com_reposicao():
    # Número total de bolas
    total_bolas = 40
    # Número de bolas azuis e roxas
    bolas_azuis_roxas = 10
    # Número de bolas a serem escolhidas
    escolhas = 10

    # Calcula o coeficiente binomial para escolher 2 bolas azuis e 2 bolas roxas
    total_maneiras = math.comb(bolas_azuis_roxas, 2) ** 2

    # Calcula a probabilidade
    probabilidade = total_maneiras / math.comb(total_bolas, escolhas)
    return probabilidade

def calcular_probabilidade_sem_reposicao():
    # Número total de bolas
    total_bolas = 40
    # Número de bolas azuis e roxas
    bolas_azuis_roxas = 10
    # Número de bolas a serem escolhidas
    escolhas = 10

    # Calcula o coeficiente binomial para escolher 2 bolas azuis e 2 bolas roxas
    total_maneiras = math.comb(bolas_azuis_roxas, 2) * math.comb(bolas_azuis_roxas, 2)

    # Calcula a probabilidade
    probabilidade = total_maneiras / math.comb(total_bolas, escolhas)
    return probabilidade

# Resultados
probabilidade_com_reposicao = calcular_probabilidade_com_reposicao()
probabilidade_sem_reposicao = calcular_probabilidade_sem_reposicao()

print(f"Probabilidade com reposição: {probabilidade_com_reposicao:.6f}")
print(f"Probabilidade sem reposição: {probabilidade_sem_reposicao:.6f}")
```

Questão 3:

Probabilidade de Ganhar:
Para ganhar, a soma dos olhos nos quatro dados deve ser inferior a 9. Logo, deve-se calcular quantas combinações possíveis resultam em uma soma menor que 9
Há 6 combinações possíveis que atendem a esse critério: (1, 1, 1, 1), (1, 1, 1, 2), (1, 1, 2, 1), (1, 2, 1, 1), (2, 1, 1, 1) e (2, 2, 1, 1).
A probabilidade de ganhar é, portanto, 6/6^4 = 1/216.

Probabilidade de Perder:
Para perder, a soma dos olhos nos quatro dados deve ser 9 ou mais.
Existem 6 combinações possíveis que atendem a esse critério: (3, 3, 3, 3), (3, 3, 3, 4), (3, 3, 4, 3), (3, 4, 3, 3), (4, 3, 3, 3) e (4, 4, 3, 3).
A probabilidade de perder é, portanto, 6/6^4 = 1/216.

Resultado Final:
Se ganhar, receberá de volta r reais (no caso, r = 10).
Se perder, perderá 1 real.
A diferença entre a probabilidade de ganhar e a probabilidade de perder é pequena (1/216 para ambos os casos).
Portanto, a longo prazo, espera-se que não ganhe nem perca dinheiro, contudo em um número de amostras muito grande, o resultado tende a ir para o negativo, demonstrando que a casa sempre vence.

```python
# Questão 3

r = 10  # Valor do retorno se a soma for inferior a 9
prob = [0] * 21  # Probabilidades de cada soma de 4 a 24
for i in range(1, 7):
    for j in range(1, 7):
        for k in range(1, 7):
            for l in range(1, 7):
                soma = i + j + k + l
                if soma < 9:
                    prob[soma - 4] += 1
total_simulations = 6**4  # Total de simulações possíveis
prob = [count / total_simulations for count in prob]

expected_value = 0
for i in range(0, len(prob)):
    if i < 5:
        expected_value += prob[i] * r
    else:
        expected_value -= prob[i]

print("Valor esperado:", expected_value)
```

```python
# Questão 4 c)

# Monte Carlo
import random
import numpy as np

# Função a ser integrada: x / (1 + x^2)^2
def func(x):
    return x / (1 + x**2)**2

# Monte Carlo
def monte_carlo_integration(func, N, a, b):
    total = 0
    for _ in range(N):
        x = random.uniform(a, b)
        total += func(x)
    return total * (b - a) / N

# Definindo os limites de integração e o número de pontos aleatórios
a_monte_carlo = 0
b_monte_carlo = 10  # Estendendo o intervalo de integração para cobrir o infinito
N_monte_carlo = 1000000

integral_estimate_monte_carlo = monte_carlo_integration(func, N_monte_carlo, a_monte_carlo, b_monte_carlo)
print("Estimativa da integral por Monte Carlo:", integral_estimate_monte_carlo)

# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

# Integração por importância
def importance_integration(func, pdf, N, a, b):
    total = 0
    for _ in range(N):
        x = random.uniform(a, b)
        total += func(x) / pdf(x)
    return total / N

# Função de densidade de probabilidade (pdf)
def pdf(x):
    return 0.5 * np.exp(-x / 2)

# Definindo os limites de integração e o número de pontos aleatórios
a_importance = 0
b_importance = 10  # Estendendo o intervalo de integração para cobrir o infinito
N_importance = 1000000

integral_estimate_importance_sampling = importance_integration(func, pdf, N_importance, a_importance, b_importance)
print("Estimativa da integral por importância:", integral_estimate_importance_sampling)
```
