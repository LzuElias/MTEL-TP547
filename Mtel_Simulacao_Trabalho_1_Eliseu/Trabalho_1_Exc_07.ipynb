```python
import numpy as np
import matplotlib.pyplot as plt

# Função de densidade de probabilidade (PDF)
def pdf(x):
    return np.exp(x) / (np.exp(2) - 1)

# Função de distribuição acumulada (CDF)
def cdf(x):
    return (np.exp(x) - 1) / (np.exp(2) - 1)

# Função inversa da CDF
def inverse_cdf(p):
    return np.log(1 + p * (np.exp(2) - 1))

# Gerar amostras aleatórias utilizando o método da transformada inversa
def generate_samples(n):
    u = np.random.rand(n)  # Amostras uniformemente distribuídas entre 0 e 1
    return inverse_cdf(u)

# Parâmetros
n_samples = 10000

# Gerar amostras
samples = generate_samples(n_samples)

# Plotagem do histograma das amostras geradas
plt.hist(samples, bins=50, density=True)
# Plotagem da função de densidade de probabilidade (PDF)
x_values = np.linspace(0, 2, 1000)
plt.plot(x_values, pdf(x_values), 'r')
plt.xlabel("")
plt.ylabel("")
plt.show()



```

```python
from ctypes import alignment
import numpy as np
import matplotlib.pyplot as plt

# Limites
a = 0
b = 2
N = 10000 # Número de amostras

# Gerar amostras aleatórias utilizando o método da transformada inversa
amt = [] # amostras
for i in range(N):
    u = np.random.rand()  # Amostra aleatória uniformemente distribuída entre 0 e 1
    x = np.log(1 + u * (np.exp(b) - 1))  # Aplicação da inversa da CDF
    amt.append(x)


# Função de densidade de probabilidade (PDF)
x_values = np.linspace(a, b, 1000)
pdf = np.exp(x_values) / (np.exp(b) - 1)

# Plotagem da função de densidade de probabilidade (PDF) e histograma
plt.hist(amt, bins=50, density=True)
plt.plot(x_values, pdf, 'r')
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```
