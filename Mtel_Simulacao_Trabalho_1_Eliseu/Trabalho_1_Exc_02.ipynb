```python
import numpy as np



lambda1=6 #Número médio de requisições
N=100000 #Numero de amostras
value=8
count=0
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)

for poisvalue in av:
    if poisvalue<=value:
        count=count+1
prob=count/N
print("a probabilidade e",prob)

print()
print(av)
```
