```python
import numpy as np
import matplotlib.pyplot as plt

M = 1.5  # M >= max(f(x) / g(x)), onde f(x) é a função alvo e g(x) é a função de envelopamento

# Número de amostras
N = 10000
amt = []

# Gerar amostras utilizando o método da aceitação/rejeição
for i in range(N):
    amt_aceitas = False
    while not amt_aceitas:
        x = np.random.uniform(-1, 1)  # Amostra uniforme da g(x)
        u = np.random.uniform(0, M)  # Amostra uniforme para aceitação/rejeição

        if u <= 1.5 * x**2:  # Comparando com a f(x)
            amt.append(x)
            amt_aceitas = True

# Função alvo
x_values = np.linspace(-1, 1, 1000)
funcao = 1.5 * x_values**2

# Plotagem da função alvo (PDF)
plt.hist(amt, bins=50, density=True)
plt.plot(x_values, funcao, 'r')
plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')
plt.title('Amostras geradas com o Método da Aceitação/Rejeição')
plt.show()
```
