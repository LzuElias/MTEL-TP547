```python
import numpy as np
import matplotlib.pyplot as plt


N=10000
lambda1=1/28
x=np.random.uniform(0,1,N)
value=4
Xexp=-np.log(x)/lambda1
count=0
print(Xexp)


for expvalue in Xexp:
    if expvalue<=value:
        count=count+1
prob=count/N
print("a probabilidade e",prob)

Probana=1-np.exp(-lambda1*value)
print("a probabilidade analítica e",Probana)

print(Xexp)
X=np.arange(0, 10, 0.1)
fx=lambda1*np.exp(-lambda1*X)
plt.plot(X,fx)
plt.hist(Xexp,bins=100,density=True)
plt.show()
```
