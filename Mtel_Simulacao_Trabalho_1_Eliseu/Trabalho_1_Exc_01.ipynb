```python
import numpy as np
import matplotlib.pyplot as plt

xa=4
x1a=np.array([xa])
na=20
aa=5
ca=0
ma=7
for ia in range(na):
    xa=(aa*xa+ca)%ma
    x1a=np.append(x1a,xa)
print(x1a)
inda=np.arange(na+1)
plt.bar(inda, x1a)
plt.show()

#%%---------------------------------------------

xb=7
x1b=np.array([xb])
nb=20
ab=5
cb=0
mb=7
for ib in range(nb):
    xb=(ab*xb+cb)%mb
    x1b=np.append(x1b,xb)
print(x1b)
indb=np.arange(nb+1)
plt.bar(indb, x1b)
plt.show()
```
