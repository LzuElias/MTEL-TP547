```python
import numpy as np
import matplotlib.pyplot as plt


#%% letra a)
lambda1 = 3 #Número médio de requisições
N=100000 #Numero de amostras
value=2
count=0
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)

for poisvalue in av:
    if poisvalue >= value:
        count=count+1
prob=count/N
print("a probabilidade e",prob)

print()
print(av)

plt.hist(av, bins=range(int(max(av))+1), density=True, alpha=0.7, color='blue', label='Histograma de av')
plt.xlabel('Valor')
plt.ylabel('Frequência Relativa')
plt.title('Histograma dos Dados Gerados (av)')
plt.grid(False)
plt.legend()
plt.show()
```
