```python
import numpy as np
import matplotlib.pyplot as plt


#%% letra a)
lambda1=0.15 #Número médio de requisições
N=100000 #Numero de amostras
value=2
count=0
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)

for poisvalue in av:
    if poisvalue<=value:
        count=count+1
prob=count/N
print("a probabilidade e",prob)

print()
print(av)

plt.figure(figsize=(10, 5))
plt.hist(av, bins=range(int(max(av))+1), density=True, alpha=0.7, color='blue', label='Histograma de av')
plt.xlabel('Valor')
plt.ylabel('Frequência Relativa')
plt.title('Histograma dos Dados Gerados (av)')
plt.grid(False)
plt.legend()
plt.show()

#%% Letra b)
lambda2=0.15 #Número médio de requisições
N2=1000 #Numero de amostras
value2=6
count2=0
av2=np.array([])
x2=np.random.uniform(0,1,N2)
for ix2 in x2:
    i2 = 0
    pr2 = np.exp(-lambda2)
    F2=pr2
    while ix2>=F2:
        pr2=lambda2/(i2+1)*pr2
        F2 = F2 + pr2
        i2 = i2 + 1;
    a2=i2
    av2=np.append(av2,a2)

for poisvalue2 in av2:
    if poisvalue2 >= value2:
        count2=count2+1
prob2=count2/N2
print("\na probabilidade e",prob2)

print()
#print(av2)

# Histograma dos dados gerados em av2
plt.figure(figsize=(10, 5))
plt.hist(av2, bins=range(int(max(av2))+1), density=True, alpha=0.7, color='red', label='Histograma de av2')
plt.xlabel('Valor')
plt.ylabel('Frequência Relativa')
plt.title('Histograma dos Dados Gerados (av2)')
plt.grid(True)
plt.legend()
plt.show()
```
